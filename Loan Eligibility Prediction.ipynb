# Importing the Dependency

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# Data Collection and preprocessing

```python
loan_dataset=pd.read_csv('Downloads/Loan_status.csv')
```

# Statistical Measure

# Number of missing values in each column

```python
loan_dataset.isnull().sum()
```

# Dropping missing value

```python
loan_dataset=loan_dataset.dropna()
```

```python
loan_dataset.isnull().sum()
```

```python
loan_dataset.head()
```

# Label Encoding:Means changing alphabetecial value to numerical value

```python
 pd.set_option("mode.copy_on_write", True)
 pd.options.mode.copy_on_write = True
```

```python
loan_dataset.replace({"Loan_Status":{'N':0,'Y':1}},inplace=True)
#loan_dataset.loc[:, "Loan_Status"].replace({"N": 0, "Y": 1}, inplace=True)
```

```python
loan_dataset.head()
```

```python
# Dependent column value
loan_dataset['Dependents'].value_counts()
```

```python
# Replacing the value of 3+ to 4
loan_dataset=loan_dataset.replace(to_replace='3+',value=4)
```

```python
loan_dataset['Dependents'].value_counts()
```

# Data Visualization

```python
#Education and Loan status
sns.countplot(x='Education',hue='Loan_Status',data=loan_dataset)
```

```python
#Convert cateogrical columns to numerical columns
loan_dataset.replace({"Married":{'No':0,'Yes':1},"Gender":{'Male':1,'Female':0},"Self_Employed":{'No':0,'Yes':1},"Property_Area":{'Rural':0,'Semiurban':1,'Urban':2},"Education":{'Graduate':1,'Not Graduate':0}},inplace=True)
```

```python
loan_dataset.head()
```

# Seperating features and target data

```python
X=loan_dataset.drop(columns=['Loan_ID','Loan_Status'],axis=1)
Y=loan_dataset['Loan_Status']
```

```python
print(X)
```

```python
print(Y)
```

# Training and splitting data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Training the model using Vector Machine Model

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

# Model Evaluation

```python
#Accuracy score on training data
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy of training data:' , training_data_accuracy)
```
